```python
import cv2 # Open CV library for various image processing
import numpy as np
import matplotlib.image as img
import matplotlib.pyplot as plt
```

```python
class Image_Calculator:
    def __init__(self, img1, img2, h, w):
        self.img1 = img1
        self.img2 = img2
        img1 = cv2.resize(img1, (w, h), interpolation=cv2.INTER_NEAREST)
        img2 = cv2.resize(img2, (w, h), interpolation=cv2.INTER_NEAREST)
        
        
    def add(self, ch1, ch2):
        
        img1_ch = img1[:, :, ch1]
        img2_ch = img2[:, :, ch2]
        output_img = img1_ch + img2_ch
        return output_img
    
    def subtract(self, ch1, ch2):
        img1_ch = img1[:, :, ch1]
        img2_ch = img2[:, :, ch2]
        output_img = img1_ch - img2_ch      
        return output_img
    
    def multiply(self, nb, ch, factor):
        if nb == 1:
            img_ch = img1[:, :, ch]
        elif nb == 2:
            img_ch = img2[:, :, ch]
        output_img = img_ch * factor
        return output_img
    
    def divide(self, nb, ch, factor):
        if nb == 1:
            img_ch = img1[:, :, ch]
        elif nb == 2:
            img_ch = img2[:, :, ch]
        output_img = img_ch * factor
        return output_img
```
